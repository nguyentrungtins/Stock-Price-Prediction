```python
from tslearn.preprocessing import TimeSeriesScalerMeanVariance
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
class Drawer:
    def create_before_normalize_line_chart(self, dataFrame):
        # Get all column names, which is stock symbol names
        columnName = list(dataFrame.columns.values)
        
        plt.figure(figsize=(20, 10))
        
        # Draw all closing prices according to each stock symbol names in first five stock symbols.
        # The closing prices are collected from 2019 to 2022 as requested. 
        plt.plot(columnName[0], columnName[1], data=dataFrame, color='green', linewidth=2)
        plt.plot(columnName[0], columnName[2], data=dataFrame, color='blue', linewidth=2)
        plt.plot(columnName[0], columnName[3], data=dataFrame, color='pink', linewidth=2)
        plt.plot(columnName[0], columnName[4], data=dataFrame, color='orange', linewidth=2)
        plt.plot(columnName[0], columnName[5], data=dataFrame, color='gold', linewidth=2)
        
        # show legend
        plt.legend()
        
        # #show title, x label, y label
        plt.title("BEFORE NORMALIZATION")
        plt.ylabel('Closing prices')
        plt.xlabel('Dates in 2020, 2021, 2022')
        
        # show graph
        plt.show()
        
    def normalizeData(self, data):
        normalize = TimeSeriesScalerMeanVariance(mu=0., std=1.).fit_transform([data])
        normalizeData = [i[0] for i in normalize[0]]
        return normalizeData        
        
    def create_after_normalize_line_chart(self, dataFrame):
        columnName = list(dataFrame.columns.values)
        
        plt.figure(figsize=(20, 10))
        
        for i in range(1, 100):
            dataFrame[columnName[i]] = self.normalizeData(dataFrame[columnName[i]])
       
        plt.plot(columnName[0], columnName[1], data=dataFrame, color='green', linewidth=2)
        plt.plot(columnName[0], columnName[2], data=dataFrame, color='blue', linewidth=2)
        plt.plot(columnName[0], columnName[3], data=dataFrame, color='pink', linewidth=2)
        plt.plot(columnName[0], columnName[4], data=dataFrame, color='orange', linewidth=2)
        plt.plot(columnName[0], columnName[5], data=dataFrame, color='gold', linewidth=2)
        
        # show legend
        plt.legend()
        
        #show title, x label, y label
        plt.title("AFTER NORMALIZATION")
        plt.ylabel('Closing prices')
        plt.xlabel('Dates in 2020, 2021, 2022')
        
        # show graph
        plt.show()
```

```python
closingPriceDataFrame = pd.read_csv("ClosingPrice.csv")

dataFrame = closingPriceDataFrame.copy()

drawer = Drawer()
drawer.create_before_normalize_line_chart(dataFrame)
drawer.create_after_normalize_line_chart(dataFrame)
```
